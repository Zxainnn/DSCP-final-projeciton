```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
print(os.getcwd())
os.chdir('C:/Users/User/OneDrive/桌面/DSCP final projeciton')
print("目前位置：", os.getcwd())
df = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
sns.lmplot(
    data=df,                       
    x='Hours_Studied',             
    y='Exam_Score',                 
    hue='Parental_Education_Level',           
    palette='Set2',                
    height=6,                       
    aspect=1.5,                   
    scatter_kws={'s': 50,'alpha':0.4}
)

plt.title('Hours Studied vs Exam Score (Shaded by Parental Education Level)')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```

```python
df['Parental_Involvement'] = df['Parental_Involvement'].map({'Low':-1,'Medium':0, 'High':1})
df['Access_to_Resources'] = df['Access_to_Resources'].map({'Low':-1,'Medium':0, 'High':1})
df['Extracurricular_Activities'] = df['Extracurricular_Activities'].map({'No':0, 'Yes':1})
df['Motivation_Level'] = df['Motivation_Level'].map({'Low':-1,'Medium':0, 'High':1})
df['Internet_Access'] = df['Internet_Access'].map({'No':0, 'Yes':1})
df['Family_Income'] = df['Family_Income'].map({'Low':-1,'Medium':0, 'High':1})
df['Teacher_Quality'] = df['Teacher_Quality'].map({'Low':-1,'Medium':0, 'High':1})
df['School_Type'] = df['School_Type'].map({'Private':0, 'Public':1})
df['Peer_Influence'] = df['Peer_Influence'].map({'Negative':-1,'Neutral':0, 'Positive':1})
df['Learning_Disabilities'] = df['Learning_Disabilities'].map({'No':0, 'Yes':1})
df['Parental_Education_Level'] = df['Parental_Education_Level'].map({'High School':-1,'College':0, 'Postgraduate':1})
df['Distance_from_Home'] = df['Distance_from_Home'].map({'Far':-1,'Moderate':0, 'Near':1})
df['Gender'] = df['Gender'].map({'Male':0, 'Female':1})
```

```python
import seaborn as sns
plt.figure(figsize=(20, 10))

sns.heatmap(df.corr(), annot=True, cmap='vlag', linewidths=0.5)
plt.title("Seaborn-Heatmap")
plt.show()
```

```python
Parental_Involvement = df['Parental_Involvement'].value_counts()
original_labels = Parental_Involvement.index.tolist()
sizes = Parental_Involvement.values

label_map = {
    -1: 'Low',
    0: 'Medium',
    1: 'High'
}

custom_labels = [label_map[label] for label in original_labels]


plt.figure(figsize=(6, 6))
plt.pie(
    sizes,                    
    labels=custom_labels,   
    autopct='%1.1f%%',      
    startangle=140,           
    colors=plt.cm.Set2.colors 
)

plt.title('Parental Involvement')
plt.axis('equal')  
plt.show()
```
